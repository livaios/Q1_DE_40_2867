**1. The relationship between the penguin's bill length and bill depth for each species.**

```python
import numpy as np
import pandas as pd
import seaborn as sns
penguins = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/penguins.csv")
sns.scatterplot(data=penguins,x=penguins["bill_length_mm"], y=penguins["bill_depth_mm"],hue=penguins["species"])
```

 **What do you observe from the graph?**

Your answer here:

*   using the bill depth and bill length we can use a classifier to classify penguins into species





**Why did you choose this type of graph?**

Your answer here:

*   given a big range of data, this has proven to be the best way to view the relation of 2 attributes given a certain class

**2. The mean BMI of female and male penguins.** 


```python
penguins_bmi = penguins.groupby(['sex'])["body_mass_g"]
sns.barplot(x='sex', y='body_mass_g',data=penguins,estimator=np.mean)
```

 **What do you observe from the graph?**

Your answer here:

*   Male penguins have higher BMI than female penguins


**Why did you choose this type of graph?**

Your answer here:

*   Because it is the best at grouping and comparing data
